```python
import os
import SimpleITK as sitk
import numpy as np
import glob
import re 
def images_to_nifti(image_folder, output_file):
    image_list= []
    for filename in sorted(os.listdir(image_folder)):
        if filename.lower().endswith(('.jpg','.png')):
            image_path = os.path.join(image_folder, filename)
            img = sitk.ReadImage(image_path)
            img_data = sitk.GetArrayFromImage(img)
            image_list.append(img_data)

    # convert the list of image arrays into a 3D array
    image_stack = np.stack(image_list, axis=-1)
                
                
    # Create a SimpleITK image object
    image_stack_sitk = sitk.GetImageFromArray(image_stack)

    #save the NIFTI image to a file 
    sitk.WriteImage(image_stack_sitk, output_file)
    
input_folder = r'D:\Local_Data_V4\LocalDataV4\HemorrhagicStroke\79\Scan'
output_file = r'D:\Local_Data_V4\LDV4_1\Hemorrhagic'
images_to_nifti(input_folder, output_nii_file)
            
```
